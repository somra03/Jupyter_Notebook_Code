# Violin Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.legend_handler import HandlerTuple
```

```python
var = sns.load_dataset("tips")
```

```python
v = sns.violinplot(x = "day",y = "total_bill",data = var,hue = "time"
                  ,legend = False,palette = "Accent")
v.set_xticklabels(v.get_xticklabels(),rotation =90)
v.set_xticklabels(v.get_xticklabels(),size = 20)
v.set_xlabel(v.get_xlabel(),size = 20)
v.set_ylabel(v.get_ylabel(),size = 20)
v.set_yticklabels(v.get_yticklabels(),size = 20)
v.legend().set_visible(False)
plt.show()
```

```python
v = sns.violinplot(x = "day",y = "total_bill",data = var
                  ,palette = "Accent")
v.set_xticklabels(v.get_xticklabels(),rotation =90)
v.set_xticklabels(v.get_xticklabels(),size = 20)
v.set_xlabel(v.get_xlabel(),size = 20)
v.set_ylabel(v.get_ylabel(),size = 20)
v.set_yticklabels(v.get_yticklabels(),size = 20)
v.legend().set_visible(False)
plt.show()
```

```python
v = sns.violinplot(x = "time",y = "total_bill",data = var
                   ,order = ["Dinner","Lunch"],saturation = 0.9
                  ,legend = False,palette = "Accent")

v.legend().set_visible(False)
plt.show()
```

```python
v = sns.violinplot(x = "day",y = "total_bill",data = var,hue = "sex"
                ,split =True,color = "r",scale = "width" )
v.set_xticklabels(v.get_xticklabels(),rotation =90)
v.set_xticklabels(v.get_xticklabels(),size = 20)
v.set_xlabel(v.get_xlabel(),size = 20)
v.set_ylabel(v.get_ylabel(),size = 20)
v.set_yticklabels(v.get_yticklabels(),size = 20)
v.legend().set_visible(False)
plt.show()
```

```python
v = sns.violinplot(x ="total_bill",y = "day",data = var,hue = "sex"
                ,split =True,color = "r",scale = "width",inner = "point" )
v.set_xticklabels(v.get_xticklabels(),rotation =90)
v.set_xticklabels(v.get_xticklabels(),size = 20)
v.set_xlabel(v.get_xlabel(),size = 20)
v.set_ylabel(v.get_ylabel(),size = 20)
v.set_yticklabels(v.get_yticklabels(),size = 20)
v.legend().set_visible(False)
plt.show()
```

# Multiple plots // Facet Grid

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("tips")
var
```

```python
fg = sns.FacetGrid(var,col="sex",hue = "day")

fg.map(plt.scatter,"total_bill","tip").add_legend()

sns.despine()

# for ax in fg.axes.flat:
#     ax.grid(False) for disabling grid function

plt.show()
```

```python
fg = sns.FacetGrid(var,col="day",hue = "sex")

fg.map(plt.bar,"total_bill","tip").add_legend()

sns.despine()

# for ax in fg.axes.flat:
#     ax.grid(False) for disabling grid function

plt.show()
```
