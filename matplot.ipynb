# Scatter Plot


```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [3,2,4,5,1,6,2]
colour = [10,40,30,29,56,20,30]
c =  ["r","b","black","violet","y","g","brown"]
sizes = [190,290,250,130,230,170,150]
plt.scatter(day,no,cmap ="viridis",c =colour,s = sizes,alpha =0.4,marker ="*",edgecolor ="g",linewidth = 2)
plt.scatter(day,no2,c = "r",s = sizes,alpha =0.4,marker ="^",edgecolor ="g",linewidth = 2)
t = plt.colorbar()
t.set_label("colorbar")
plt.title("Scatter Plot",fontsize= 15)
plt.xlabel("Day",fontsize= 15)
plt.ylabel("No",fontsize= 15)
plt.show()
```

# Stack Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels = l)
plt.legend()

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
#baseline = "wiggle"
#baseline = "sym"
```

# Step plot

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y,color = "violet",marker = "o",ms =10, mfc ="b",label = "python")
plt.grid()

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()

plt.show()
```

# Fill_Between Plot

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y,color = "grey")

plt.fill_between(x,y,color = "g",where= (x>=2) & (x<=4),alpha = 0.2)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
